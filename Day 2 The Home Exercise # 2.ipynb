# Create your own Pandemic Analyzer

```python
from tkinter import *
import tkinter


#Declaration
window=Tk()
window.resizable(0, 0)
window.configure(background="light grey")
window.geometry("400x400")
window.title("Pandemic Analytics Engine")



def cal_func():

    total_population = entry2.get()
    int_total_population = int(total_population)

    infected = entry3.get()
    int_infected = int(infected)

    death = entry4.get()
    int_death = int(death)

    
    
    
    
    #formula used
    ifr = (int_death / int_infected) * 100
    cmr = (int_death / int_total_population)*100000 #resident deaths per 100,000 population in pirticular area
    
    
    
    window = Tk()
    window.title("Pandemic Analyzer")
    window.configure(background="light grey")
    window.geometry("400x400")
    
    
    
    label1 = Label(window, font=("Arial", 18, "bold"), text="Coronavirus disease" , bg = "light grey" , fg = "dark blue")
    label1.grid(row=0, column=0)

    label2 = Label(window, font=("Arial", 14, "bold"), text="IFR : ", bg = "light grey" , fg = "dark blue")
    label2.grid(row=2, column=0)
   
    label3 = Label(window, text=ifr)
    label3.grid(row=2, column=1)

    label4 = Label(window, font=("Arial", 14, "bold"), text="CMR : ", bg = "light grey" , fg = "dark blue")
    label4.grid(row=3, column=0)
    
    label5 = Label(window, text=cmr)
    label5.grid(row=3, column=1)
    label6= Label(window, font=("Arial", 8),text="(resident deaths per 100,000 population in pirticular area)" ,bg="light grey", fg = "dark blue")
    label6.grid(row=5,column=0)
    
    
# Listing out all the states:    
options = ['Andaman and Nicobar', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telengana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
select = StringVar() 
select.set(options[0])




# font and colour.
label1 = Label(window,font=("Arial", 18, "bold"), text="Pandemic Analyzer",bg = "light grey" , fg = "dark blue")
label1.grid(row=0,column= 1)

label2 = Label(window, text="Select State",bg = "light green" , fg = "dark blue")
label2.grid(row=2,column=0)
drop = OptionMenu(window,select,*options)
drop.grid(row=2,column=1)

label3 = Label(window, text="population:" ,bg="light grey" , fg = "dark blue")
label3.grid(row=3,column=0)
entry2 = Entry(window)
entry2.grid(row=3, column=1)

label4 = Label(window, text="infected" ,bg="light grey", fg = "dark blue")
label4.grid(row=4,column=0)
entry3 = Entry(window)
entry3.grid(row=4, column=1)

label5 = Label(window, text="decerased" ,bg="light grey", fg = "dark blue")
label5.grid(row=5,column=0)
entry4 = Entry(window)
entry4.grid(row=5, column=1)

label6 = Label(window, font=("Arial", 10, "bold"),text="IFR(Infection Fatality Rate)" ,bg="light grey", fg = "dark blue")
label6.grid(row=10,column=0)
label7= Label(window, font=("Arial", 10, "bold"),text="CMR(Curde Mortality Rate)" ,bg="light grey", fg = "dark blue")
label7.grid(row=15,column=0)


# calculating:
button1 = Button(window, text="   IFR    ",command=cal_func , bg = "grey" , fg = "dark blue")
button1.grid(row=7,column=1)
button1 = Button(window, text="  CMR  ",command=cal_func , bg = "grey" , fg = "dark blue")
button1.grid(row=8,column=1)

# Displaying the GUI on to the console:
window.mainloop()



```
