# Create your own Pandemic Analyzer

```python
from tkinter import *
from tkinter.ttk import Combobox
from tkinter import messagebox as mbox
from csv import reader
window=Tk()
window.resizable(0, 0)
window.configure(background="light grey")
window.geometry("280x300")
window.title("Pandemic Analytics Engine")
with open('covid19.csv', 'r') as obj:

    data = reader(obj)
    x = list(data)
    a=[]
    b=[]
#Creating List using Loop
for i in range(len(x)):
    a.append(x[i])
    b.append(a[i][0])
# Creating Function for clearing the text boxes
def enable_items():
    infected.configure(state='normal')
    recovered.configure(state='normal')
    deceased.configure(state='normal')
    population.configure(state='normal')
    
    infected.delete(0, 'end')
    recovered.delete(0, 'end')
    deceased.delete(0, 'end')
    population.delete(0, 'end')  
    
    CMRRes.configure(state='normal')
    CMRRes.delete(0, 'end')
    
    IFRRes.configure(state='normal')
    IFRRes.delete(0, 'end')
  





  ##Creating Function for population data into text boxes and disabling it

def on_state_changed():
    #print(data1)
    enable_items()
    for i in range(len(a)):
            #print(data1)
    
            dat=a[i][0]
            if cb.get()==dat:
                infected.insert(END,a[i][1])
                recovered.insert(END,a[i][2])
                deceased.insert(END,a[i][3])
                population.insert(END,a[i][4])
                infected.configure(state='disabled')
                recovered.configure(state='disabled')
                deceased.configure(state='disabled')
                population.configure(state='disabled')

                

def ifr():                                      
        k=int(infected.get())
        l=int(recovered.get())
        m=int(deceased.get())
        n=int(population.get())
        result=round((m/k)*100,2)
        IFRRes.insert(END, str(result))
        IFRRes.configure(state='disabled')
      

    
# cmr=(deaths/population)*100000
def cmr():                           
        k=int(infected.get())
        l=int(recovered.get())
        m=int(deceased.get())
        n=int(population.get())
        result=round((m/n)*100000,2)
        CMRRes.insert(END, str(result)) 
        CMRRes.configure(state='disabled')
     

          
label1 = Label(window,font=("Arial", 18, "bold"), text="Pandemic Analyzer",bg = "light grey" , fg = "dark blue")
label1.grid(row=0,column= 0)
label2 = Label(window, text="Select State",bg = "light green" , fg = "dark blue")
label2.place(x=5, y=35,width=70)
cb=Combobox(window,values=b)
cb.place(x=100, y=35,width=125,height=20)
label4 = Label(window, text="infected" ,bg="light grey", fg = "red")
label4.place(x=5, y=55,width=70)
infected=Entry()
infected.place(x=100, y=55,width=125,height=20)
label4 = Label(window, text="recoved" ,bg="light grey", fg = "red")
label4.place(x=5, y=75,width=70)
recovered=Entry()
recovered.place(x=100, y=75,width=125,height=20)
label5 = Label(window, text="decerased" ,bg="light grey", fg = "red")
label5.place(x=5, y=95,width=70)
deceased=Entry()
deceased.place(x=100, y=95,width=125,height=20)
label3 = Label(window, text="population" ,bg="light grey" , fg = "red")
label3.place(x=5, y=115,width=70)
population=Entry()
population.place(x=100, y=115,width=125,height=20)
chb=Checkbutton(window, text = "confirm",command=on_state_changed)
chb.place(x=125,y=150)
IFR = Button(window, text=' IFR ',bg="snow",fg="red",command=ifr)
IFR.place(x=5, y=200,width=70)
CMR=Button(window, text='CMR',bg="snow",fg="red",command=cmr)
CMR.place(x=5, y=240,width=70)
IFRRes=Entry()
IFRRes.place(x=100, y=200,width=122,height=20)
CMRRes=Entry()
CMRRes.place(x=100, y=240,width=122,height=20)
window.mainloop()
```
